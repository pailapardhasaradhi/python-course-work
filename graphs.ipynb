1. Line Plot(plt.plot)
Matplotlib is a Python library for data visualization.

It helps you turn numbers into graphs and charts so data is easier to understand.

Commonly used in data analysis, machine learning, reporting, and presentations.

Without visualization, itâ€™s hard to see patterns in raw data (tables, CSV files, etc.).

ðŸ‘‰ In short: Matplotlib = data â†’ picture â†’ insights.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30, 6, 8, 12, 15, 18, 20, 22, 25, 30])
y = np.array([2, 5, 70, 11, 13, 16, 18, 90, 22, 27])

plt.plot(x, y, color='blue', linestyle='--', linewidth=5,
         marker='o', markersize=8, markerfacecolor='green')

plt.xlabel("----X Axis---")
plt.ylabel("---Y Axis")
plt.title("Line Plot Example")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Bar Chart(plt.bar)
Concept: Uses rectangular bars to compare categories.

Use: Good for comparisons between groups.

Example: Sales of products (A, B, C), population in cities.

```python
import matplotlib.pyplot as plt

categories = ['laptop','perfume','smartphone','watch']
sales = [50,70,90,20]

plt.bar(categories, sales, color='yellow', edgecolor='green',
        linewidth=3)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

3. Histogram(plt.hist)
Concept: Divides data into bins (intervals) and shows frequency.

Use: Helps see distribution of data (normal, skewed, etc.).

Example: Age distribution of students, exam score frequency.

```python
import matplotlib.pyplot as plt
import numpy as np

data =  np.array([2,4,3,1,6,5, 70, 11, 13, 16, 18, 90, 22, 27])
plt.hist(data, bins=20, color='cyan', edgecolor='black', alpha=0.8)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

4. Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 100, 50)
colors = np.random.rand(50)
plt.scatter(x, y, s=sizes, c=colors, cmap='plasma',
            alpha=0.8, edgecolors='black')
plt.colorbar(label="Color Scale")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot Example")
plt.show()
```

5. Pie Chart
Concept: Circle divided into slices showing proportions.

Use: Visualize percentage/share of categories.

Example: Market share of companies, budget allocation.

```python
import matplotlib.pyplot as plt

batches = ['PFS-30', 'DS-15', 'DA-14', 'PFS-31']
sizes = [53, 16, 18, 17]
colors = ['red', 'yellow', 'pink', 'brown']

plt.pie(sizes, labels=batches, colors=colors, autopct='%1.2f%%',
        startangle=90, explode=[0, 0, 0.1, 0])
plt.title("Pie Chart Example")
plt.show()
```

6. Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.randn(100),
        np.random.randn(100) * 2,
        np.random.randn(100) * 1.5]

plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'],
            patch_artist=True,
            boxprops=dict(facecolor="lightblue"))

plt.title("Box Plot Example")
plt.show()




```

7. Heatmap (Seaborn)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.rand(10, 10)
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap Example")
plt.show()
```

8. 3D Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c='red', marker='o')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Scatter Plot Example")
plt.show()
```

9. Subplots (Multiple Plots in One Figure)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]
data = np.random.randn(100)

fig, axs = plt.subplots(2, 2, figsize=(8, 8))

axs[0, 0].plot(x, y, 'b--')
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(categories, values, color='purple')
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(np.random.rand(20), np.random.rand(20), color='green')
axs[1, 0].set_title("Scatter Plot")

axs[1, 1].hist(data, bins=20, color='orange')
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
values = [10, 20, 15, 30]
categories = ['A', 'B', 'C', 'D']
data = np.random.randn(100)

fig, axs = plt.subplots(2, 3, figsize=(12, 6))

# Line plot
axs[0, 0].plot(x, y, 'b--')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(categories, values, color='purple')
axs[0, 1].set_title("Bar Chart")

# Histogram
axs[0, 2].hist(data, bins=20, color='orange')
axs[0, 2].set_title("Histogram")

# Scatter plot
axs[1, 0].scatter(np.random.rand(20), np.random.rand(20), color='green')
axs[1, 0].set_title("Scatter Plot")

# Pie chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")

# Area chart
y1 = np.linspace(1, 10, 10)
y2 = np.linspace(2, 20, 10)
axs[1, 2].fill_between(y1, y2, color="skyblue", alpha=0.5)
axs[1, 2].set_title("Area Chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

x = np.linspace(0, 10, 100)
y = np.sin(x)
values = [10, 20, 15, 30]
categories = ['A', 'B', 'C', 'D']
data = np.random.randn(100)

fig, axs = plt.subplots(3, 3, figsize=(12, 12))

# Line plot
axs[0, 0].plot(x, y, 'r')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(categories, values, color='purple')
axs[0, 1].set_title("Bar Chart")

# Histogram
axs[0, 2].hist(data, bins=15, color='orange')
axs[0, 2].set_title("Histogram")

# Scatter
axs[1, 0].scatter(np.random.rand(20), np.random.rand(20), color='green')
axs[1, 0].set_title("Scatter Plot")

# Pie
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")

# Area
axs[1, 2].fill_between(np.arange(1, 11), np.linspace(2, 20, 10), color="skyblue", alpha=0.5)
axs[1, 2].set_title("Area Chart")

# Box
axs[2, 0].boxplot([np.random.randn(100), np.random.randn(100)*2])
axs[2, 0].set_title("Box Plot")

# Heatmap (in subplot using seaborn)
sns.heatmap(np.random.rand(5,5), ax=axs[2,1], cmap="coolwarm", cbar=False)
axs[2, 1].set_title("Heatmap")

# Empty 3D placeholder
axs[2, 2].text(0.5, 0.5, "3D Plot\n(run separately)", ha='center', va='center')
axs[2, 2].set_xticks([])
axs[2, 2].set_yticks([])

plt.tight_layout()
plt.show()
```

10. Area Chart


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, 10)
y1 = np.array([3, 6, 8, 12, 15, 18, 20, 22, 25, 30])
y2 = np.array([2, 5, 7, 10, 13, 16, 18, 19, 22, 27])

plt.fill_between(x, y1, color="blue", alpha=0.5, label="Dataset 1")
plt.fill_between(x, y2, color="green", alpha=0.5, label="Dataset 2")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Area Chart Example")
plt.legend()
plt.grid(True)
plt.show()
```
