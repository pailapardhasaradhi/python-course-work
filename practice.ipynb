```python
import matplotlib.pyplot as mp
x = [23,45,67,58,43,21]
y = [45,67,89,54,32,12]
mp.plot(x, y, 
         color='blue',       # line color
         linestyle='--',     # dashed line
         linewidth=5,        # line thickness
         marker='o',         # circle markers
         markersize=8,       # marker size
         markerfacecolor='green') 
mp.xlabel("pardhu")
mp.ylabel("paila")
mp.grid()
mp.show()
```

```python
import matplotlib.pyplot as mp
days = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
temperature = [30,45,30,36,21,40,50]
mp.plot(days,temperature,color = 'pink',linestyle = '--',linewidth = 5,marker = 's',markerfacecolor = 'blue')
mp.xlabel('-----days-----')
mp.ylabel('-----temp in celsius----')
mp.grid()
mp.title("----TEMPERTURE ANALYSIS----")
mp.show()
```
